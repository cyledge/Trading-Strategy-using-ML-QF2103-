```python
#Test for staionarity
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)

    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)

test_stationarity(MSFT)

# https://www.kaggle.com/code/nageshsingh/stock-market-forecasting-arima
```

```python
# To separate the trend and the seasonality from a time series,
# we can decompose the series using the following code.
result = seasonal_decompose(MSFT, model='multiplicative', period = 30)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(16, 9)
```

```python
# Split data into train and test sets
X = MSFT.values
size = int(len(X) * 0.8)
train, test = X[0:size], X[size:len(X)]
```

## the Auto ARIMA model provided the optimal parameters p,d, and q as 0, 1 and 0 respectively

```python
model_autoARIMA = auto_arima(train, start_p=1, start_q=1,test='adf',max_p=4, max_q=4,
                      m=2,d=None,seasonal=False,start_P=1,D=2,trace=True,error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)
print(model_autoARIMA.summary())
model_autoARIMA.plot_diagnostics(figsize=(15,8))
plt.show()

# https://www.kaggle.com/code/nageshsingh/stock-market-forecasting-arima/notebook
```

```python
# Define the ARIMA model
def arima_forecast(history):
    # Fit the model
    model = ARIMA(history, order=(0,1,1))
    model_fit = model.fit()

    # Make the prediction
    output = model_fit.forecast()
    yhat = output[0]
    return yhat

# https://medium.com/shikhars-data-science-projects/predicting-stock-prices-using-arima-fourier-transformation-and-deep-learning-e5fb4f693c85
```

```python
# Walk-forward validation
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    # Generate a prediction
    yhat = arima_forecast(history)
    predictions.append(yhat)
    # Add the predicted value to the training set
    obs = test[t]
    history.append(obs)

# https://medium.com/shikhars-data-science-projects/predicting-stock-prices-using-arima-fourier-transformation-and-deep-learning-e5fb4f693c85
```

```python
plt.figure(figsize=(12, 6), dpi=100)
plt.plot(MSFT.iloc[size:,:].index, test, label='Real')
plt.plot(MSFT.iloc[size:,:].index, predictions, color='red', label='Predicted')
plt.title('ARIMA Predictions vs Actual Values')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# https://medium.com/shikhars-data-science-projects/predicting-stock-prices-using-arima-fourier-transformation-and-deep-learning-e5fb4f693c85
```

```python
np.mean(np.abs(predictions - test)/np.abs(test))
```

```python
MSFT_test = MSFT.iloc[size:len(X)]
MSFT_test.loc[:,'Ari_Pred_Price'] = predictions # stock price predicted by ARIMA
# Benchmark returns
MSFT_test['Returns'] = np.log(MSFT_test['MSFT.O']/MSFT_test['MSFT.O'].shift(1))
# Returns predicted by Arima
MSFT_test['Ari_pred_ret'] = np.log(MSFT_test['Ari_Pred_Price']/MSFT_test['Ari_Pred_Price'].shift(1))
# direction predicted by Arima
MSFT_test['Ari_pred_dir'] = np.sign(MSFT_test['Ari_Pred_Price']-MSFT_test['MSFT.O'].shift(1))
MSFT_test.dropna(inplace = True)
# long when predicted direction is +1, vice versa, then compute returns
MSFT_test['Ari_ret'] = MSFT_test['Ari_pred_dir'] * MSFT_test['Returns']
MSFT_test
```

```python
# Compare total returns from arima with benchmark
MSFT_test[['Returns', 'Ari_ret']].sum().apply(np.exp)
```

```python
# plot the cumulative returns
MSFT_test[['Returns', 'Ari_ret']].cumsum().apply(np.exp).plot()
plt.title('MSFT')
```

```python
raw = pd.read_csv('tr_eikon_eod_data.csv',
                 index_col = 0, parse_dates = True)
raw.info()
```

```python
vol = pd.DataFrame()
raw1 = yf.download(['AAPL', 'MSFT', 'INTC', 'AMZN', 'GS'], start = '2010-01-01', end= '2018-06-30')
vol['AAPL.O'] = raw1['Volume']['AAPL']
vol['MSFT.O'] = raw1['Volume']['MSFT']
vol['INTC.O'] = raw1['Volume']['INTC']
vol['AMZN.O'] = raw1['Volume']['AMZN']
vol['GS.N'] = raw1['Volume']['GS']
vol
```

```python
data = raw[['AAPL.O', 'MSFT.O', 'INTC.O', 'AMZN.O', 'GS.N']]
data.dropna(inplace=True)
data
```

```python
EWM = data.ewm(alpha=0.75 , min_periods = 20 , adjust = False).mean()
EWM
```

```python
delta = data.diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)
average_gain = gain.rolling(window=14, min_periods=14).mean()
average_loss = loss.rolling(window=14, min_periods=14).mean()
RS = average_gain / average_loss
RSI = 100 - (100 / (1 + RS))
RSI
```

```python
rets = np.log(data / data.shift(1))
rets
```

```python
rets['AAPLvola'] = rets['AAPL.O'].rolling(14).std()
rets['INTCvola'] = rets['INTC.O'].rolling(14).std()
rets['MSFTvola'] = rets['MSFT.O'].rolling(14).std()
rets['AMZNvola'] = rets['AMZN.O'].rolling(14).std()
rets['GSvola'] = rets['GS.N'].rolling(14).std()
rets['AAPLvolat'] = rets['AAPL.O'].rolling(42).std()
rets['INTCvolat'] = rets['INTC.O'].rolling(42).std()
rets['MSFTvolat'] = rets['MSFT.O'].rolling(42).std()
rets['AMZNvolat'] = rets['AMZN.O'].rolling(42).std()
rets['GSvolat'] = rets['GS.N'].rolling(42).std()
rets
```

```python
data['RSI_Signal_AAPL.O'] = np.where(RSI['AAPL.O'] < 30, 1, np.where(RSI['AAPL.O'] > 70, -1, 0))
data['RSI_Signal_MSFT.O'] = np.where(RSI['MSFT.O'] < 30, 1, np.where(RSI['MSFT.O'] > 70, -1, 0))
data['RSI_Signal_INTC.O'] = np.where(RSI['INTC.O'] < 30, 1, np.where(RSI['INTC.O'] > 70, -1, 0))
data['RSI_Signal_AMZN.O'] = np.where(RSI['AMZN.O'] < 30, 1, np.where(RSI['AMZN.O'] > 70, -1, 0))
data['RSI_Signal_GS.N'] = np.where(RSI['GS.N'] < 30, 1, np.where(RSI['GS.N'] > 70, -1, 0))
data
```

```python
MSFT = data[['MSFT.O']]
```

```python
MSFT = pd.merge(MSFT, EWM['MSFT.O'], left_index=True, right_index=True, how='left')
MSFT = MSFT.rename(columns={'MSFT.O_x': 'close', "MSFT.O_y": "EWM"})
MSFT
```

```python
MSFT = pd.merge(MSFT, RSI['MSFT.O'], left_index=True, right_index=True, how='left')
MSFT = MSFT.rename(columns={"MSFT.O": "RSI"})
MSFT['return'] = np.log(MSFT['close'] / MSFT['close'].shift(1))
MSFT
```

```python
MSFT['PnL'] = np.where(MSFT['return'] > 0, 1, -1)
MSFT
```

```python
MSFT['RSI lag1'] = MSFT['RSI'].shift(1)
MSFT['RSI bin'] = pd.cut(MSFT['RSI'], bins=20, labels = False)
MSFT['RSI bin'] += 1
MSFT['RSI bin lag1'] = MSFT['RSI bin'].shift(1)
MSFT.dropna(inplace = True)
MSFT
```

```python
for i in range(1,6):
    colname = "lag" + str(i)
    MSFT[colname] =MSFT['return'].shift(i)

MSFT['EWM_normal'] = (MSFT['EWM'] - MSFT['EWM'].min()) / (MSFT['EWM'].max() - MSFT['EWM'].min())
MSFT['RSI normal'] = (MSFT['RSI'] - MSFT['RSI'].min()) / (MSFT['RSI'].max() - MSFT['RSI'].min())
MSFT
```

```python
MSFT['ewa lag1'] = MSFT['EWM_normal'].shift(1)
MSFT['RSI lag1'] = MSFT['RSI normal'].shift(1)
MSFT['close lag1'] = MSFT['close'].shift(1)
MSFT['SPYlag1'] = raw['SPY'].shift(1)
MSFT['VIXlag1'] = raw['.VIX'].shift(1)
MSFT['EURlag1'] = raw['EUR='].shift(1)
MSFT['GLDlag1'] = raw['GLD'].shift(1)
MSFT[['AAPLvollag1','AMZNvollag1','GSvollag1','INTCvollag1','MSFTvollag1']] = vol[['AAPL.O','AMZN.O','GS.N','INTC.O','MSFT.O' ]].shift(1)
MSFT[['AAPLlag1','AMZNlag1','GSlag1','INTClag1','MSFTlag1']] = data[['AAPL.O','AMZN.O','GS.N','INTC.O','MSFT.O' ]].shift(1)
MSFT['volalag1'] =rets['MSFTvola'].shift(1)
MSFT.dropna(inplace = True)
MSFT
```

```python
MSFT_neu_lag = MSFT.copy() # to prevent from modifying original dataset
MSFT_neu_lag['neu_pred'] = 0
MSFT_neu_lag
```

```python
X_train = MSFT_neu_lag[['volalag1','AAPLlag1','AMZNlag1','GSlag1','INTClag1','MSFTlag1', 'AAPLvollag1','AMZNvollag1','GSvollag1','INTCvollag1','MSFTvollag1','ewa lag1', 'lag1', 'lag2', 'RSI lag1','SPYlag1','VIXlag1','EURlag1','GLDlag1']][:1628]
X_test = MSFT_neu_lag[['volalag1','AAPLlag1','AMZNlag1','GSlag1','INTClag1','MSFTlag1', 'AAPLvollag1','AMZNvollag1','GSvollag1','INTCvollag1','MSFTvollag1', 'ewa lag1', 'lag1', 'lag2', 'RSI lag1','SPYlag1','VIXlag1','EURlag1','GLDlag1']][1628:]
pca = PCA(n_components = 8)  # Choose number of components, 2 is the only one that give an acceptable return
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

model = LogisticRegression()
MSFT_neu_lag['neu_pred'][1628:] = model.fit(X_train_pca, MSFT_neu_lag['PnL'][:1628]).predict(X_test_pca)

(MSFT_neu_lag['neu_pred'] * MSFT_neu_lag['return'])[1628:].cumsum().apply(np.exp).plot()
```

```python
MSFT_test['neu_pred'] = MSFT_neu_lag['neu_pred'][1628:]
MSFT_test['volalag1'] = MSFT_neu_lag['volalag1'][1628:]
MSFT_test['volatlag1'] = rets['MSFTvolat'].shift(1)
MSFT_test['modpred'] = np.where(MSFT_test['volalag1'] > MSFT_test['volatlag1'],MSFT_test['neu_pred'],  MSFT_test['Ari_pred_dir'])
MSFT_test
```

```python
MSFT_test['modifiedreturn']=MSFT_test['modpred'] * MSFT_test['Returns']
MSFT_test['arimareturn']=MSFT_test['Ari_pred_dir'] * MSFT_test['Returns']
MSFT_test['Returns'].cumsum().apply(np.exp).plot()
MSFT_test['arimareturn'].cumsum().apply(np.exp).plot()
MSFT_test['modifiedreturn'].cumsum().apply(np.exp).plot()
plt.legend()
plt.title('MSFT')
```
